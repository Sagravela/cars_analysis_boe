```python
import camelot
from tqdm import tqdm
import pandas as pd
```

```python
df = pd.DataFrame()

for page in tqdm(range(3, 604), desc="Extracting tables"):
    table = camelot.read_pdf('BOE-A-2016-11948.pdf', pages=str(page+1))[0].df
    table = table.iloc[1:,:]
    df = pd.concat([df, table], axis=0)
```

```python
columns = ['brand', 'model/kind', 'business_period', 'c_c', 'n_c', 'g_d', 'p_kw', 'cvf', 'co2_gkm', 'cv', 'value_2017_eu']
df.columns = columns
df['cvf'] = df['cvf'].replace(",", ".", regex=True)
df = df.apply(pd.to_numeric, downcast='integer', errors='ignore')
df = df.dropna(how='all')
df.reset_index(drop=True, inplace=True)
df
```

```python
df.to_csv('cars.csv', index=False)
```
